```sql
USE CATALOG tibia_analytics;
USE SCHEMA silver;
```

```sql
CREATE TABLE IF NOT EXISTS silver.worlds_summary (
  snapshot_id STRING,
  players_online INT,
  record_players INT,
  record_at TIMESTAMP,
  api_version INT,
  api_release STRING,
  api_commit STRING,
  api_processed_at TIMESTAMP,
  tibia_urls ARRAY<STRING>,
  error INT,
  http_code INT,
  message STRING,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.worlds_summary IS 'Summary of Tibia Worlds.';
COMMENT ON COLUMN silver.worlds_summary.snapshot_id IS 'Technical unique identifier.';
COMMENT ON COLUMN silver.worlds_summary.players_online IS 'The total number of currently online players across all worlds.';
COMMENT ON COLUMN silver.worlds_summary.record_players IS 'The highest total number of players online across all worlds.';
COMMENT ON COLUMN silver.worlds_summary.record_at IS 'The timestamp when the highest number of players online was recorded (UTC).'; 
COMMENT ON COLUMN silver.worlds_summary.api_version IS 'The API major version currently running.';
COMMENT ON COLUMN silver.worlds_summary.api_release IS 'The API release currently running.';
COMMENT ON COLUMN silver.worlds_summary.api_commit IS 'The API GitHub commit SHA.';
COMMENT ON COLUMN silver.worlds_summary.api_processed_at IS 'The timestamp from when the data was processed (UTC).';
COMMENT ON COLUMN silver.worlds_summary.tibia_urls IS 'The links to the sources of the data on tibia.com.';
COMMENT ON COLUMN silver.worlds_summary.error IS 'The error code thrown by TibiaData API for identification of issue.';
COMMENT ON COLUMN silver.worlds_summary.http_code IS 'The HTTP response code from the API.';
COMMENT ON COLUMN silver.worlds_summary.message IS 'The error message thrown by TibiaData API for human readability.';
COMMENT ON COLUMN silver.worlds_summary.source_file_date IS 'Reference date of the source data snapshot.';
COMMENT ON COLUMN silver.worlds_summary.source_file_path IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN silver.worlds_summary.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN silver.worlds_summary.ingested_at IS 'Bronze layer ingestion timestamp (UTC).';
COMMENT ON COLUMN silver.worlds_summary.processed_at IS 'Silver layer processing timestamp (UTC).';
```

```sql
CREATE TABLE IF NOT EXISTS silver.worlds (
  snapshot_id STRING,
  world_id STRING,
  world STRING,
  status STRING,
  players_online INT,
  location STRING,
  pvp_type STRING,
  premium_only BOOLEAN,
  transfer_type STRING,
  battleye_protected BOOLEAN,
  battleye_date DATE,
  world_type STRING,
  tournament_type STRING,
  source_file_date DATE,
  ingested_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.worlds IS 'Detailed information about Tibia Worlds.';
COMMENT ON COLUMN silver.worlds.snapshot_id IS 'Technical unique identifier.';
COMMENT ON COLUMN silver.worlds.world_id IS 'World unique identifier.';
COMMENT ON COLUMN silver.worlds.world IS 'The name of the world.';
COMMENT ON COLUMN silver.worlds.status IS 'The current operational status of the world. Possible values: Online, Offline.';
COMMENT ON COLUMN silver.worlds.players_online IS 'The number of currently online players in the world.';
COMMENT ON COLUMN silver.worlds.location IS 'The physical location of the servers. Possible values: Europe, North America, Oceania, South America.';
COMMENT ON COLUMN silver.worlds.pvp_type IS 'The type of PvP. Possible values: Hardcore PvP, Open PvP, Optional PvP, Retro Hardcore PvP, Retro Open PvP.';
COMMENT ON COLUMN silver.worlds.premium_only IS 'Indicates whether only premium account players are allowed to play on the world.';
COMMENT ON COLUMN silver.worlds.transfer_type IS 'The type of transfer restriction applied to the world. Possible values: Blocked, Locked, Regular.';
COMMENT ON COLUMN silver.worlds.battleye_protected IS 'Indicates whether the world is protected by BattlEye. TRUE if protected, FALSE otherwise.';
COMMENT ON COLUMN silver.worlds.battleye_date IS 'The date when BattlEye was added. If battleye_protected is TRUE, NULL indicates protection since world launch.';
COMMENT ON COLUMN silver.worlds.world_type IS 'The game world category. Possible values: Experimental, Regular, Tournament.';
COMMENT ON COLUMN silver.worlds.tournament_type IS 'The tournament world category. Possible values: Default, Regular, Restricted.';
COMMENT ON COLUMN silver.worlds.source_file_date IS 'Reference date of the source data snapshot.';
COMMENT ON COLUMN silver.worlds.ingested_at IS 'Bronze layer ingestion timestamp (UTC).';
COMMENT ON COLUMN silver.worlds.processed_at IS 'Silver layer processing timestamp (UTC).';
```
