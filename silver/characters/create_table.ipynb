```sql
USE CATALOG tibia_analytics;
USE SCHEMA silver;
```

```sql
CREATE TABLE IF NOT EXISTS silver.characters_summary (
  snapshot_id STRING,
  snapshot_character_id STRING,
  name STRING,
  api_version INT,
  api_release STRING,
  api_commit STRING,
  api_processed_at TIMESTAMP,
  tibia_urls ARRAY<STRING>,
  error INT,
  http_code INT,
  message STRING,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.characters_summary IS 'Summary of Tibia Characters.';
COMMENT ON COLUMN silver.characters_summary.snapshot_id IS 'Technical unique identifier.';
COMMENT ON COLUMN silver.characters_summary.snapshot_character_id IS 'Character unique identifier.';
COMMENT ON COLUMN silver.characters_summary.name IS 'The name of the character.';
COMMENT ON COLUMN silver.characters_summary.api_version IS 'The API major version currently running.';
COMMENT ON COLUMN silver.characters_summary.api_release IS 'The API release currently running.';
COMMENT ON COLUMN silver.characters_summary.api_commit IS 'The API GitHub commit SHA.';
COMMENT ON COLUMN silver.characters_summary.api_processed_at IS 'The timestamp from when the data was processed (UTC).';
COMMENT ON COLUMN silver.characters_summary.tibia_urls IS 'The links to the sources of the data on tibia.com.';
COMMENT ON COLUMN silver.characters_summary.error IS 'The error code thrown by TibiaData API for identification of issue.';
COMMENT ON COLUMN silver.characters_summary.http_code IS 'The HTTP response code from the API.';
COMMENT ON COLUMN silver.characters_summary.message IS 'The error message thrown by TibiaData API for human readability.';
COMMENT ON COLUMN silver.characters_summary.source_file_date IS 'Reference date of the source data snapshot.';
COMMENT ON COLUMN silver.characters_summary.source_file_path IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN silver.characters_summary.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN silver.characters_summary.ingested_at IS 'Bronze layer ingestion timestamp (UTC).';
COMMENT ON COLUMN silver.characters_summary.processed_at IS 'Silver layer processing timestamp (UTC).';
```

```sql
CREATE TABLE IF NOT EXISTS silver.characters_history (
  snapshot_id STRING,
  snapshot_character_id STRING,
  name STRING,
  traded BOOLEAN,
  deletion_at TIMESTAMP,
  former_names ARRAY<STRING>,
  title STRING,
  unlocked_titles INT,
  sex STRING,
  vocation STRING,
  level INT,
  achievement_points INT,
  world_id STRING,
  world STRING,
  former_worlds ARRAY<STRING>,
  residence STRING,
  married_to STRING,
  houses ARRAY<STRUCT<
    house_id STRING,
    name STRING,
    town STRING,
    paid_until_date DATE
  >>,
  guild STRUCT<
    name STRING,
    rank STRING
  >,
  last_login_at TIMESTAMP,
  comment STRING,
  position STRING,
  account_status STRING,
  account_badges ARRAY<STRUCT<
    name STRING,
    description STRING,
    icon_url STRING
  >>,
  achievements ARRAY<STRUCT<
    grade INT,
    name STRING,
    secret BOOLEAN
  >>,
  deaths_truncated BOOLEAN,
  deaths ARRAY<STRUCT<
    death_at TIMESTAMP,
    level INT,
    reason STRING,
    assists ARRAY<STRUCT<
      name STRING,
      player BOOLEAN,
      traded BOOLEAN,
      summon STRING
    >>,
    killers ARRAY<STRUCT<
      name STRING,
      player BOOLEAN,
      traded BOOLEAN,
      summon STRING
    >>
  >>,
  account_information STRUCT<
    position STRING,
    loyalty_title STRING,
    created_at TIMESTAMP
  >,
  other_characters ARRAY<STRUCT<
    name STRING,
    position STRING,
    world STRING,
    status STRING,
    main BOOLEAN,
    traded BOOLEAN,
    deleted BOOLEAN
  >>,
  source_file_date DATE,
  api_timestamp TIMESTAMP,
  ingested_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.characters_history IS 'Summary of Tibia Characters.';
COMMENT ON COLUMN silver.characters_history.snapshot_id IS 'Technical unique identifier.';
COMMENT ON COLUMN silver.characters_history.snapshot_character_id IS 'Character unique identifier.';
COMMENT ON COLUMN silver.characters_history.name IS 'The name of the character.';
COMMENT ON COLUMN silver.characters_history.traded IS 'Whether the character was traded. Last 6 months.';
COMMENT ON COLUMN silver.characters_history.deletion_at IS 'The timestamp when the character will be deleted (UTC), if scheduled for deletion.';
COMMENT ON COLUMN silver.characters_history.former_names IS 'List of former names of the character.';
COMMENT ON COLUMN silver.characters_history.title IS 'The character\'s selected title.';
COMMENT ON COLUMN silver.characters_history.unlocked_titles IS 'The number of titles the character has unlocked.';
COMMENT ON COLUMN silver.characters_history.sex IS 'The character\'s sex.';
COMMENT ON COLUMN silver.characters_history.vocation IS 'The character\'s vocation.';
COMMENT ON COLUMN silver.characters_history.level IS 'The character\'s level.';
COMMENT ON COLUMN silver.characters_history.achievement_points IS 'The total of achievement points the character has.';
COMMENT ON COLUMN silver.characters_history.world_id IS 'World unique identifier.';
COMMENT ON COLUMN silver.characters_history.world IS 'The character\'s current world.';
COMMENT ON COLUMN silver.characters_history.former_worlds IS 'List of former worlds the character was in. Last 6 months.';
COMMENT ON COLUMN silver.characters_history.residence IS 'The character\'s current residence.';
COMMENT ON COLUMN silver.characters_history.married_to IS 'The name of the character\'s husband/spouse.';
COMMENT ON COLUMN silver.characters_history.houses IS 'List of houses the character owns currently.';
COMMENT ON COLUMN silver.characters_history.guild IS 'The guild that the character is member of.';
COMMENT ON COLUMN silver.characters_history.last_login_at IS 'The character\'s last logged in time (UTC).';
COMMENT ON COLUMN silver.characters_history.comment IS 'The character\'s comment.';
COMMENT ON COLUMN silver.characters_history.position IS 'The character\'s special position.';
COMMENT ON COLUMN silver.characters_history.account_status IS 'Whether account is Free or Premium.';
COMMENT ON COLUMN silver.characters_history.account_badges IS 'The account\'s badges.';
COMMENT ON COLUMN silver.characters_history.achievements IS 'The character\'s achievements.';
COMMENT ON COLUMN silver.characters_history.deaths_truncated IS 'Whether the character\'s deaths were truncated or not.';
COMMENT ON COLUMN silver.characters_history.deaths IS 'The character\'s deaths.';
COMMENT ON COLUMN silver.characters_history.account_information IS 'The account information.';
COMMENT ON COLUMN silver.characters_history.other_characters IS 'The account\'s other characters.';
COMMENT ON COLUMN silver.characters_history.source_file_date IS 'Reference date of the source data snapshot.';
COMMENT ON COLUMN silver.characters_history.api_timestamp IS 'API capture timestamp (UTC).';
COMMENT ON COLUMN silver.characters_history.ingested_at IS 'Bronze layer ingestion timestamp (UTC).';
COMMENT ON COLUMN silver.characters_history.processed_at IS 'Silver layer processing timestamp (UTC).';
```

```sql
CREATE TABLE IF NOT EXISTS silver.characters_identity (
  character_id STRING,
  current_name STRING,
  current_world STRING,
  locked_names ARRAY<STRING>,
  all_names ARRAY<STRING>,
  first_seen_date DATE,
  last_seen_date DATE,
  deleted_at TIMESTAMP,
  is_deleted BOOLEAN,
  created_at TIMESTAMP,
  updated_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.characters_identity IS 'Consolidated identity table for Tibia characters.';
COMMENT ON COLUMN silver.characters_identity.character_id IS 'Stable unique identifier of the character across name changes.';
COMMENT ON COLUMN silver.characters_identity.current_name IS 'The current active name of the character at the latest snapshot.';
COMMENT ON COLUMN silver.characters_identity.current_world IS 'The current active world of the character at the latest snapshot.';
COMMENT ON COLUMN silver.characters_identity.locked_names IS 'Array of former names currently locked by the game and not available for reuse.';
COMMENT ON COLUMN silver.characters_identity.all_names IS 'Array containing the complete historical list of names associated with the character.';
COMMENT ON COLUMN silver.characters_identity.first_seen_date IS 'Earliest snapshot date in which the character was observed.';
COMMENT ON COLUMN silver.characters_identity.last_seen_date IS 'Most recent snapshot date in which the character was observed';
COMMENT ON COLUMN silver.characters_identity.deleted_at IS 'Timestamp when the character was effectively deleted according to the scheduled deletion date provided by the API (UTC).';
COMMENT ON COLUMN silver.characters_identity.is_deleted IS 'Indicates whether the character is considered deleted based on the scheduled deletion date provided by the API.';
COMMENT ON COLUMN silver.characters_identity.created_at IS 'Record creation timestamp (UTC).';
COMMENT ON COLUMN silver.characters_identity.updated_at IS 'Last record update timestamp (UTC).';
COMMENT ON COLUMN silver.characters_identity.processed_at IS 'Silver layer processing timestamp (UTC).';
```

```sql

CREATE TABLE IF NOT EXISTS silver.characters_state (
  character_id STRING,
  name STRING,
  world STRING,
  name_type STRING,
  valid_from DATE,
  valid_to DATE,
  is_active_lock BOOLEAN,
  created_at TIMESTAMP,
  updated_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.characters_state IS 'Character name state history table. Tracks temporal status of each name associated with a character.';
COMMENT ON COLUMN silver.characters_state.character_id IS 'Stable unique identifier of the character across name changes.';
COMMENT ON COLUMN silver.characters_state.name IS 'The name of the character.';
COMMENT ON COLUMN silver.characters_state.world IS 'The world of the character.';
COMMENT ON COLUMN silver.characters_state.name_type IS 'The current classification of the name. Possible values: observed_current, observed_historical, inferred_historical.';
COMMENT ON COLUMN silver.characters_state.valid_from IS 'Start snapshot date in which the character was observed.';
COMMENT ON COLUMN silver.characters_state.valid_to IS 'End snapshot date in which the character was observed with same attributes.';
COMMENT ON COLUMN silver.characters_state.is_active_lock IS 'Indicates whether the name is currently locked and unavailable for reuse. IF name_type is ''inferred_historical'', THEN is_active_lock must be FALSE.';
COMMENT ON COLUMN silver.characters_state.created_at IS 'Record creation timestamp (UTC).';
COMMENT ON COLUMN silver.characters_state.updated_at IS 'Last record update timestamp (UTC).';
COMMENT ON COLUMN silver.characters_state.processed_at IS 'Silver layer processing timestamp (UTC).';
```

```sql

CREATE TABLE IF NOT EXISTS tibia_analytics.silver.characters_attempts (
  character_id STRING,
  name STRING,
  world STRING,
  failure_count INT,
  last_status STRING,
  last_snapshot_date DATE,
  created_at TIMESTAMP,
  updated_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.characters_attempts IS 'Consolidated state of character API fetch attempts with consecutive failure tracking.';
COMMENT ON COLUMN silver.characters_attempts.character_id IS 'Stable unique identifier of the character across name changes.';
COMMENT ON COLUMN silver.characters_attempts.name IS 'The name of the character.';
COMMENT ON COLUMN silver.characters_attempts.world IS 'The world of the character.';
COMMENT ON COLUMN silver.characters_attempts.failure_count IS 'Number of consecutive failed API attempts for the same character lifecycle.';
COMMENT ON COLUMN silver.characters_attempts.last_status IS 'Status of the last API attempt. Possible values: success, failed.';
COMMENT ON COLUMN silver.characters_attempts.last_snapshot_date IS 'Business snapshot date of the last processed API attempt.';
COMMENT ON COLUMN silver.characters_attempts.created_at IS 'Record creation timestamp (UTC).';
COMMENT ON COLUMN silver.characters_attempts.updated_at IS 'Last record update timestamp (UTC).';
COMMENT ON COLUMN silver.characters_attempts.processed_at IS 'Silver layer processing timestamp (UTC).';
```
