```sql
USE CATALOG tibia_analytics;
USE SCHEMA utils;
```

```sql
CREATE TABLE IF NOT EXISTS utils.calendar (
  date_key              INT,
  year_key              INT,
  quarter_key           INT,
  month_key             INT,
  week_key              INT,
  full_date             DATE,
  year                  INT,
  quarter               INT,
  quarter_name          STRING,
  quarter_year          STRING,
  quarter_start_date    DATE,
  quarter_end_date      DATE,
  month                 INT,
  month_name            STRING,
  month_short_name      STRING,
  month_year_short      STRING,
  month_start_date      DATE,
  month_end_date        DATE,
  days_in_month         INT,
  iso_week              INT,
  iso_year              INT,
  iso_year_week         INT,
  week_start_date       DATE,
  week_end_date         DATE,
  day                   INT,
  day_of_year           INT,
  day_of_week_iso       INT,
  day_name              STRING,
  day_short_name        STRING,
  is_first_day_of_month BOOLEAN,
  is_last_day_of_month  BOOLEAN,
  is_weekday            BOOLEAN,
  is_weekend            BOOLEAN,
  processed_at          TIMESTAMP
);
```

```sql
COMMENT ON TABLE utils.calendar IS 'Date dimension used across Tibia Analytics. Provides calendar attributes, period keys, labels and temporal flags to support filtering, grouping and time-based analysis. One row per calendar day.';
COMMENT ON COLUMN utils.calendar.date_key IS 'Unique identifier for the calendar date in YYYYMMDD format.';
COMMENT ON COLUMN utils.calendar.year_key IS 'Calendar year identifier.';
COMMENT ON COLUMN utils.calendar.quarter_key IS 'Calendar quarter identifier in YYYYQQ format.';
COMMENT ON COLUMN utils.calendar.month_key IS 'Calendar month identifier in YYYYMM format.';
COMMENT ON COLUMN utils.calendar.week_key IS 'Calendar week identifier based on week start date.';
COMMENT ON COLUMN utils.calendar.full_date IS 'Calendar date.';
COMMENT ON COLUMN utils.calendar.year IS 'Calendar year.';
COMMENT ON COLUMN utils.calendar.quarter IS 'Calendar quarter number (1-4).';
COMMENT ON COLUMN utils.calendar.quarter_name IS 'Quarter label (Q1-Q4).';
COMMENT ON COLUMN utils.calendar.quarter_year IS 'Quarter and year label.';
COMMENT ON COLUMN utils.calendar.quarter_start_date IS 'First day of the quarter.';
COMMENT ON COLUMN utils.calendar.quarter_end_date IS 'Last day of the quarter.';
COMMENT ON COLUMN utils.calendar.month IS 'Calendar month number (1-12).';
COMMENT ON COLUMN utils.calendar.month_name IS 'Full month name.';
COMMENT ON COLUMN utils.calendar.month_short_name IS 'Abbreviated month name.';
COMMENT ON COLUMN utils.calendar.month_year_short IS 'Short month and year label.';
COMMENT ON COLUMN utils.calendar.month_start_date IS 'First day of the month.';
COMMENT ON COLUMN utils.calendar.month_end_date IS 'Last day of the month.';
COMMENT ON COLUMN utils.calendar.days_in_month IS 'Number of days in the month.';
COMMENT ON COLUMN utils.calendar.iso_week IS 'ISO week number.';
COMMENT ON COLUMN utils.calendar.iso_year IS 'ISO week-based year.';
COMMENT ON COLUMN utils.calendar.iso_year_week IS 'ISO year-week identifier.';
COMMENT ON COLUMN utils.calendar.week_start_date IS 'First day of the week.';
COMMENT ON COLUMN utils.calendar.week_end_date IS 'Last day of the week.';
COMMENT ON COLUMN utils.calendar.day IS 'Day of month.';
COMMENT ON COLUMN utils.calendar.day_of_year IS 'Day number within the year.';
COMMENT ON COLUMN utils.calendar.day_of_week_iso IS 'ISO day of week where Monday=1 and Sunday=7.';
COMMENT ON COLUMN utils.calendar.day_name IS 'Full weekday name.';
COMMENT ON COLUMN utils.calendar.day_short_name IS 'Abbreviated weekday name.';
COMMENT ON COLUMN utils.calendar.is_first_day_of_month IS 'Indicates whether the date is the first day of the month.';
COMMENT ON COLUMN utils.calendar.is_last_day_of_month IS 'Indicates whether the date is the last day of the month.';
COMMENT ON COLUMN utils.calendar.is_weekday IS 'Indicates whether the date falls on a weekday.';
COMMENT ON COLUMN utils.calendar.is_weekend IS 'Indicates whether the date falls on a weekend.';
COMMENT ON COLUMN utils.calendar.processed_at IS 'Timestamp when the record was processed into the Utils layer (UTC).';
```
