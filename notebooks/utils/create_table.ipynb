```sql
USE CATALOG tibia_analytics;
 USE SCHEMA utils;
```

```sql
CREATE TABLE IF NOT EXISTS calendar (
  -- Calendar keys
  date_key              INT,
  year_key              INT,
  quarter_key           INT,
  month_key             INT,
  week_key              INT,

  -- Calendar attributes
  full_date             DATE,
  year                  INT,
  quarter               INT,
  quarter_name          STRING,
  quarter_year          STRING,
  quarter_start_date    DATE,
  quarter_end_date      DATE,
  month                 INT,
  month_name            STRING,
  month_short_name      STRING,
  month_year_short      STRING,
  month_start_date      DATE,
  month_end_date        DATE,
  days_in_month         INT,
  iso_week              INT,
  iso_year              INT,
  iso_year_week         INT,
  week_start_date       DATE,
  week_end_date         DATE,
  day                   INT,
  day_of_year           INT,
  day_of_week_iso       INT,
  day_name              STRING,
  day_short_name        STRING,

  -- Calendar flags
  is_first_day_of_month BOOLEAN,
  is_last_day_of_month  BOOLEAN,
  is_weekday            BOOLEAN,
  is_weekend            BOOLEAN,

  -- Processing metadata
  processed_at          TIMESTAMP
);
```

```sql
COMMENT ON TABLE  calendar                       IS 'Calendar dimension maintained in the Utils layer. Provides date attributes, period identifiers, and temporal flags used across Tibia Analytics for filtering, grouping, and time-based analysis. One row per calendar day.';
COMMENT ON COLUMN calendar.date_key              IS 'Unique identifier for the calendar date in YYYYMMDD format.';
COMMENT ON COLUMN calendar.year_key              IS 'Calendar year identifier in YYYY format.';
COMMENT ON COLUMN calendar.quarter_key           IS 'Calendar quarter identifier in YYYYQQ format.';
COMMENT ON COLUMN calendar.month_key             IS 'Calendar month identifier in YYYYMM format.';
COMMENT ON COLUMN calendar.week_key              IS 'Calendar week identifier based on the week start date.';
COMMENT ON COLUMN calendar.full_date             IS 'Calendar date used as the primary reference for daily analysis.';
COMMENT ON COLUMN calendar.year                  IS 'Calendar year.';
COMMENT ON COLUMN calendar.quarter               IS 'Calendar quarter number (1-4).';
COMMENT ON COLUMN calendar.quarter_name          IS 'Quarter label (Q1-Q4).';
COMMENT ON COLUMN calendar.quarter_year          IS 'Quarter-year label.';
COMMENT ON COLUMN calendar.quarter_start_date    IS 'First day of the quarter.';
COMMENT ON COLUMN calendar.quarter_end_date      IS 'Last day of the quarter.';
COMMENT ON COLUMN calendar.month                 IS 'Month number (1-12).';
COMMENT ON COLUMN calendar.month_name            IS 'Full month name.';
COMMENT ON COLUMN calendar.month_short_name      IS 'Abbreviated month name.';
COMMENT ON COLUMN calendar.month_year_short      IS 'Short month-year label.';
COMMENT ON COLUMN calendar.month_start_date      IS 'First day of the month.';
COMMENT ON COLUMN calendar.month_end_date        IS 'Last day of the month.';
COMMENT ON COLUMN calendar.days_in_month         IS 'Number of days in the month.';
COMMENT ON COLUMN calendar.iso_week              IS 'ISO week number.';
COMMENT ON COLUMN calendar.iso_year              IS 'ISO week-based year.';
COMMENT ON COLUMN calendar.iso_year_week         IS 'ISO year-week identifier.';
COMMENT ON COLUMN calendar.week_start_date       IS 'First day of the week.';
COMMENT ON COLUMN calendar.week_end_date         IS 'Last day of the week.';
COMMENT ON COLUMN calendar.day                   IS 'Day of the month.';
COMMENT ON COLUMN calendar.day_of_year           IS 'Day number within the year.';
COMMENT ON COLUMN calendar.day_of_week_iso       IS 'ISO day of week where Monday=1 and Sunday=7.';
COMMENT ON COLUMN calendar.day_name              IS 'Full weekday name.';
COMMENT ON COLUMN calendar.day_short_name        IS 'Abbreviated weekday name.';
COMMENT ON COLUMN calendar.is_first_day_of_month IS 'Indicates whether the date is the first day of the month.';
COMMENT ON COLUMN calendar.is_last_day_of_month  IS 'Indicates whether the date is the last day of the month.';
COMMENT ON COLUMN calendar.is_weekday            IS 'Indicates whether the date falls on a weekday.';
COMMENT ON COLUMN calendar.is_weekend            IS 'Indicates whether the date falls on a weekend.';
COMMENT ON COLUMN calendar.processed_at          IS 'Timestamp when the record was processed into the Utils layer (UTC).';
```
