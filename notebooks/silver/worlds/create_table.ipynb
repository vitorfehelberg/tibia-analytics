```sql
USE CATALOG tibia_analytics;
 USE SCHEMA silver;
```

```sql
CREATE TABLE IF NOT EXISTS worlds (
  -- Snapshot metadata
  snapshot_id        STRING,

  -- World identity
  world_id           STRING,
  world              STRING,

  -- World attributes
  status             STRING,
  players_online     INT,
  location           STRING,
  pvp_type           STRING,
  premium_only       BOOLEAN,
  transfer_type      STRING,
  battleye_protected BOOLEAN,
  battleye_date      DATE,
  world_type         STRING,
  tournament_type    STRING,

  -- Processing metadata
  source_file_date   DATE,
  processed_at       TIMESTAMP
);
```

```sql
COMMENT ON TABLE  worlds                    IS 'World snapshot data collected from the TibiaData API, including status, population, location, PvP configuration, and world metadata.';
COMMENT ON COLUMN worlds.snapshot_id        IS 'Identifier of the source snapshot propagated from Bronze.';
COMMENT ON COLUMN worlds.world_id           IS 'SHA1 identifier generated from the normalized world name. Used as the stable world identifier.';
COMMENT ON COLUMN worlds.world              IS 'The name of the world.';
COMMENT ON COLUMN worlds.status             IS 'The current operational status of the world. Possible values: Online, Offline.';
COMMENT ON COLUMN worlds.players_online     IS 'The number of currently online players in the world.';
COMMENT ON COLUMN worlds.location           IS 'The physical location of the servers. Possible values: Europe, North America, Oceania, South America.';
COMMENT ON COLUMN worlds.pvp_type           IS 'The type of PvP. Possible values: Hardcore PvP, Open PvP, Optional PvP, Retro Hardcore PvP, Retro Open PvP.';
COMMENT ON COLUMN worlds.premium_only       IS 'Indicates whether only premium account players are allowed to play on the world.';
COMMENT ON COLUMN worlds.transfer_type      IS 'The type of transfer restriction applied to the world. Possible values: Blocked, Locked, Regular.';
COMMENT ON COLUMN worlds.battleye_protected IS 'Indicates whether the world is protected by BattlEye.';
COMMENT ON COLUMN worlds.battleye_date      IS 'The date when BattlEye was added. If battleye_protected is TRUE, NULL indicates protection since world launch.';
COMMENT ON COLUMN worlds.world_type         IS 'The game world category. Possible values: Experimental, Regular, Tournament.';
COMMENT ON COLUMN worlds.tournament_type    IS 'The tournament world category. Possible values: Default, Regular, Restricted.';
COMMENT ON COLUMN worlds.source_file_date   IS 'Reference date of the source snapshot extracted from the file path.';
COMMENT ON COLUMN worlds.processed_at       IS 'Timestamp when the record was processed into the Silver layer (UTC).';
```
