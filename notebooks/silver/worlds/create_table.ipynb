```sql
USE CATALOG tibia_analytics;
USE SCHEMA silver;
```

```sql
CREATE TABLE IF NOT EXISTS silver.worlds (
  snapshot_id STRING,
  world_id STRING,
  world STRING,
  status STRING,
  players_online INT,
  location STRING,
  pvp_type STRING,
  premium_only BOOLEAN,
  transfer_type STRING,
  battleye_protected BOOLEAN,
  battleye_date DATE,
  world_type STRING,
  tournament_type STRING,
  source_file_date DATE,
  ingested_at TIMESTAMP,
  processed_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE silver.worlds IS 'Detailed information about Tibia Worlds.';
COMMENT ON COLUMN silver.worlds.snapshot_id IS 'Technical unique identifier.';
COMMENT ON COLUMN silver.worlds.world_id IS 'World unique identifier.';
COMMENT ON COLUMN silver.worlds.world IS 'The name of the world.';
COMMENT ON COLUMN silver.worlds.status IS 'The current operational status of the world. Possible values: Online, Offline.';
COMMENT ON COLUMN silver.worlds.players_online IS 'The number of currently online players in the world.';
COMMENT ON COLUMN silver.worlds.location IS 'The physical location of the servers. Possible values: Europe, North America, Oceania, South America.';
COMMENT ON COLUMN silver.worlds.pvp_type IS 'The type of PvP. Possible values: Hardcore PvP, Open PvP, Optional PvP, Retro Hardcore PvP, Retro Open PvP.';
COMMENT ON COLUMN silver.worlds.premium_only IS 'Indicates whether only premium account players are allowed to play on the world.';
COMMENT ON COLUMN silver.worlds.transfer_type IS 'The type of transfer restriction applied to the world. Possible values: Blocked, Locked, Regular.';
COMMENT ON COLUMN silver.worlds.battleye_protected IS 'Indicates whether the world is protected by BattlEye. TRUE if protected, FALSE otherwise.';
COMMENT ON COLUMN silver.worlds.battleye_date IS 'The date when BattlEye was added. If battleye_protected is TRUE, NULL indicates protection since world launch.';
COMMENT ON COLUMN silver.worlds.world_type IS 'The game world category. Possible values: Experimental, Regular, Tournament.';
COMMENT ON COLUMN silver.worlds.tournament_type IS 'The tournament world category. Possible values: Default, Regular, Restricted.';
COMMENT ON COLUMN silver.worlds.source_file_date IS 'Reference date of the source data snapshot.';
COMMENT ON COLUMN silver.worlds.ingested_at IS 'Bronze layer ingestion timestamp (UTC).';
COMMENT ON COLUMN silver.worlds.processed_at IS 'Silver layer processing timestamp (UTC).';
```
