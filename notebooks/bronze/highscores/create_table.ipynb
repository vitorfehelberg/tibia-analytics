```sql
USE CATALOG tibia_analytics;
 USE SCHEMA bronze;
```

```sql
CREATE TABLE IF NOT EXISTS raw_highscores (
  -- Record identifier
  raw_highscores_id       STRING,
  
  -- API payload
  highscores              STRUCT<
    world: STRING,
    category: STRING,
    vocation: STRING,
    highscore_age: INT,
    highscore_list: ARRAY<STRUCT<
      rank: INT,
      name: STRING,
      title: STRING,
      vocation: STRING,
      world: STRING,
      level: INT,
      value: INT
    >>,
    highscore_page: STRUCT<
      current_page: INT,
      total_pages: INT,
      total_records: INT
    >
  >,

  -- API metadata
  information             STRUCT<
    api: STRUCT<
      version: INT,
      release: STRING,
      commit: STRING
    >,
    timestamp: STRING,
    tibia_urls: ARRAY<STRING>,
    status: STRUCT<
      error: INT,
      http_code: INT,
      message: STRING
    >
  >,
  
  -- Ingestion metadata
  source_file_date        DATE,
  source_file_path        STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at             TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_highscores                         IS 'Append-only Bronze table containing raw TibiaData API responses and ingestion metadata.';
COMMENT ON COLUMN raw_highscores.raw_highscores_id       IS 'SHA1 of the source file path and modification time. Stable across re-ingestion of the same unmodified file.';
COMMENT ON COLUMN raw_highscores.highscores              IS 'Raw highscores payload returned by the TibiaData API. Values are stored without business transformations.';
COMMENT ON COLUMN raw_highscores.information             IS 'API response metadata including version, timestamp, and response status information.';
COMMENT ON COLUMN raw_highscores.source_file_date        IS 'Reference date of the source snapshot extracted from the file path.';
COMMENT ON COLUMN raw_highscores.source_file_path        IS 'Source file path used for lineage and debugging purposes.';
COMMENT ON COLUMN raw_highscores.source_file_modified_at IS 'Timestamp when the source file was modified (UTC).';
COMMENT ON COLUMN raw_highscores.ingested_at             IS 'Timestamp when the record was ingested into the Bronze layer (UTC).';
```
