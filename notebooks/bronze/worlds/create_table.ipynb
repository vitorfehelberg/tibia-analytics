```sql
USE CATALOG tibia_analytics;
 USE SCHEMA bronze;
```

```sql
CREATE TABLE IF NOT EXISTS raw_worlds (
  -- Technical identifier
  raw_worlds_id           STRING,
  
  -- API payload
  worlds                  STRUCT<
    players_online: INT,
    record_players: INT,
    record_date: STRING,
    regular_worlds: ARRAY<STRUCT<
      name: STRING,
      status: STRING,
      players_online: INT,
      location: STRING,
      pvp_type: STRING,
      premium_only: BOOLEAN,
      transfer_type: STRING,
      battleye_protected: BOOLEAN,
      battleye_date: STRING,
      game_world_type: STRING,
      tournament_world_type: STRING
    >>,
    tournament_worlds: ARRAY<STRUCT<
      name: STRING,
      status: STRING,
      players_online: INT,
      location: STRING,
      pvp_type: STRING,
      premium_only: BOOLEAN,
      transfer_type: STRING,
      battleye_protected: BOOLEAN,
      battleye_date: STRING,
      game_world_type: STRING,
      tournament_world_type: STRING
    >>
  >,

  -- API metadata
  information             STRUCT<
    api: STRUCT<
      version: INT,
      release: STRING,
      commit: STRING
    >,
    timestamp: STRING,
    tibia_urls: ARRAY<STRING>,
    status: STRUCT<
      error: INT,
      http_code: INT,
      message: STRING
    >
  >,
  
  -- Ingestion metadata
  source_file_date        DATE,
  source_file_path        STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at             TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_worlds                         IS 'Append-only Bronze table containing raw TibiaData API responses and ingestion metadata.';
COMMENT ON COLUMN raw_worlds.raw_worlds_id           IS 'SHA1 of the source file path and modification time. Stable across re-ingestion of the same unmodified file. Propagated to Silver as snapshot_id.';
COMMENT ON COLUMN raw_worlds.worlds                  IS 'Raw worlds payload returned by the TibiaData API. Values are stored without business transformations.';
COMMENT ON COLUMN raw_worlds.information             IS 'API response metadata including version, timestamp, and response status information.';
COMMENT ON COLUMN raw_worlds.source_file_date        IS 'Reference date of the source snapshot extracted from the file path.';
COMMENT ON COLUMN raw_worlds.source_file_path        IS 'Source file path used for lineage and debugging purposes.';
COMMENT ON COLUMN raw_worlds.source_file_modified_at IS 'Timestamp when the source file was modified (UTC).';
COMMENT ON COLUMN raw_worlds.ingested_at             IS 'Timestamp when the record was ingested into the Bronze layer (UTC).';
```
