```sql
USE CATALOG tibia_analytics;
USE SCHEMA bronze;
```

```sql
CREATE TABLE IF NOT EXISTS raw_worlds (
  raw_worlds_id STRING,
  worlds STRUCT<
    players_online: INT,
    record_players: INT,
    record_date: STRING,
    regular_worlds: ARRAY<STRUCT<
      name: STRING,
      status: STRING,
      players_online: INT,
      location: STRING,
      pvp_type: STRING,
      premium_only: BOOLEAN,
      transfer_type: STRING,
      battleye_protected: BOOLEAN,
      battleye_date: STRING,
      game_world_type: STRING,
      tournament_world_type: STRING
    >>,
    tournament_worlds: ARRAY<STRUCT<
      name: STRING,
      status: STRING,
      players_online: INT,
      location: STRING,
      pvp_type: STRING,
      premium_only: BOOLEAN,
      transfer_type: STRING,
      battleye_protected: BOOLEAN,
      battleye_date: STRING,
      game_world_type: STRING,
      tournament_world_type: STRING
    >>
  >,
  information STRUCT<
    api: STRUCT<
      version: INT,
      release: STRING,
      commit: STRING
    >,
    timestamp: STRING,
    tibia_urls: ARRAY<STRING>,
    status: STRUCT<
      error: INT,
      http_code: INT,
      message: STRING
    >
  >,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_worlds                         IS 'Raw worlds data with ingestion metadata.';
COMMENT ON COLUMN raw_worlds.raw_worlds_id           IS 'Technical unique identifier.';
COMMENT ON COLUMN raw_worlds.worlds                  IS 'Raw worlds data extracted from the source API.';
COMMENT ON COLUMN raw_worlds.information             IS 'Source API metadata.';
COMMENT ON COLUMN raw_worlds.source_file_date        IS 'Business reference date of the source data snapshot.';
COMMENT ON COLUMN raw_worlds.source_file_path        IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN raw_worlds.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN raw_worlds.ingested_at             IS 'Bronze layer ingestion timestamp (UTC).';
```
