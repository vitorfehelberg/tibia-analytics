```sql
USE CATALOG tibia_analytics;
USE SCHEMA gold;
```

```sql
CREATE TABLE IF NOT EXISTS cohort_retention (
  granularity STRING,
  cohort_period DATE,
  cohort_period_status STRING,
  cohort_is_partial_period BOOLEAN,
  observation_period DATE,
  observation_period_status STRING,
  observation_is_partial_period BOOLEAN,
  periods_elapsed INT,
  world_id STRING,
  account_status STRING,
  sex STRING,
  vocation STRING,
  in_guild BOOLEAN,
  owns_house BOOLEAN,
  is_married BOOLEAN,
  loyalty_title STRING,
  cohort_size INT,
  retained_count INT,
  retention_rate DECIMAL(10, 6),
  processed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (granularity);
```

```sql
COMMENT ON TABLE cohort_retention IS 'Periodic cohort retention dataset used to analyze player retention behavior over time across multiple temporal granularities and segmentation dimensions. Each row represents the retained portion of a player cohort observed after a given number of elapsed periods.';
COMMENT ON COLUMN cohort_retention.granularity IS 'Temporal aggregation level used to define cohorts and retention windows. Supported values include Day, Week, Month, and Quarter.';
COMMENT ON COLUMN cohort_retention.cohort_period IS 'Start date of the cohort period in which the player first qualified for inclusion in the cohort.';
COMMENT ON COLUMN cohort_retention.cohort_period_status IS 'Data completeness status of the cohort period. Supported values include full, partial_start, partial_current, and partial_missing.';
COMMENT ON COLUMN cohort_retention.cohort_is_partial_period IS 'Indicates whether the cohort period has incomplete snapshot coverage relative to the expected number of days for the selected granularity.';
COMMENT ON COLUMN cohort_retention.observation_period IS 'Start date of the observation period used to measure retention for the cohort.';
COMMENT ON COLUMN cohort_retention.observation_period_status IS 'Data completeness status of the observation period. Supported values include full, partial_start, partial_current, and partial_missing.';
COMMENT ON COLUMN cohort_retention.observation_is_partial_period IS 'Indicates whether the observation period has incomplete snapshot coverage relative to the expected number of days for the selected granularity.';
COMMENT ON COLUMN cohort_retention.periods_elapsed IS 'Number of elapsed periods between the cohort_period and the observation_period according to the selected granularity. Example: for Month granularity, value 1 represents the first month after cohort acquisition.';
COMMENT ON COLUMN cohort_retention.world_id IS 'Stable unique identifier of the Tibia world associated with the player during the observation period.';
COMMENT ON COLUMN cohort_retention.account_status IS 'Indicates whether the player account was Free or Premium during the observation period.';
COMMENT ON COLUMN cohort_retention.sex IS 'Current sex of the character during the observation period.';
COMMENT ON COLUMN cohort_retention.vocation IS 'Normalized vocation category of the character during the observation period. Promotions are grouped into their base vocation categories.';
COMMENT ON COLUMN cohort_retention.in_guild IS 'Indicates whether the character belonged to a guild during the observation period.';
COMMENT ON COLUMN cohort_retention.owns_house IS 'Indicates whether the character owned at least one house during the observation period.';
COMMENT ON COLUMN cohort_retention.is_married IS 'Indicates whether the character was married during the observation period.';
COMMENT ON COLUMN cohort_retention.loyalty_title IS 'Loyalty title associated with the character account during the observation period.';
COMMENT ON COLUMN cohort_retention.cohort_size IS 'Total number of distinct characters included in the original cohort for the specified segmentation dimensions.';
COMMENT ON COLUMN cohort_retention.retained_count IS 'Number of distinct cohort characters that remained active in the corresponding observation_period.';
COMMENT ON COLUMN cohort_retention.retention_rate IS 'Retention ratio calculated as retained_count divided by cohort_size. Formula: retained_count / cohort_size.';
COMMENT ON COLUMN cohort_retention.processed_at IS 'Timestamp when this record was processed into the Gold layer (UTC).';
```
