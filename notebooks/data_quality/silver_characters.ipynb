```python
result = spark.sql("""
  SELECT COUNT(*) AS failed_count
    FROM tibia_analytics.silver.characters_history_enriched
   WHERE character_id IS NULL
""")

row = result.first()
missing_character_identity_count = row["failed_count"] if row is not None else 0
```

```python
check_results = {
    "Characters with missing identity": missing_character_identity_count
}
    
failed_checks = [
    f"{check_name} ({failed_count} occurrence(s))"
    for check_name, failed_count in check_results.items()
    if failed_count > 0
]

if failed_checks:
    raise Exception(
        "Silver Characters data quality checks failed:\n- "
        + "\n- ".join(failed_checks)
    )

print("All Silver Characters data quality checks passed.")
```
