```python
result = spark.sql("""
  SELECT COUNT(*) AS failed_count
    FROM tibia_analytics.silver.worlds
   WHERE source_file_date = (
           SELECT MAX(source_file_date)
             FROM tibia_analytics.silver.worlds
         )
     AND (
           world IS NULL
           OR TRIM(world) = ''
     )
""")
row = result.first()
world_name_missing_count = row["failed_count"] if row is not None else 0
```

```python
check_results = {
    "Worlds with missing names": world_name_missing_count
}

failed_checks = [
    f"{check_name} ({failed_count} occurrence(s))"
    for check_name, failed_count in check_results.items()
    if failed_count > 0
]

if failed_checks:
    raise Exception(
        "Silver Worlds data quality checks failed:\n- "
        + "\n- ".join(failed_checks)
    )

print("All Silver Worlds data quality checks passed.")
```
