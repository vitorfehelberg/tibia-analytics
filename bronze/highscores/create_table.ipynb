```sql
USE CATALOG tibia_analytics;
USE SCHEMA bronze;
```

```sql
CREATE TABLE IF NOT EXISTS raw_highscores (
  raw_highscores_id STRING,
  highscores STRUCT<
    world: STRING,
    category: STRING,
    vocation: STRING,
    highscore_age: INT,
    highscore_list: ARRAY<STRUCT<
      rank: INT,
      name: STRING,
      title: STRING,
      vocation: STRING,
      world: STRING,
      level: INT,
      value: INT
    >>,
    highscore_page: STRUCT<
      current_page: INT,
      total_pages: INT,
      total_records: INT
    >
  >,
  information STRUCT<
    api: STRUCT<
      version: INT,
      release: STRING,
      commit: STRING
    >,
    timestamp: STRING,
    tibia_urls: ARRAY<STRING>,
    status: STRUCT<
      error: INT,
      http_code: INT,
      message: STRING
    >
  >,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_highscores                         IS 'Raw highscores data with ingestion metadata.';
COMMENT ON COLUMN raw_highscores.raw_highscores_id       IS 'Technical unique identifier.';
COMMENT ON COLUMN raw_highscores.highscores              IS 'Raw highscores data extracted from the source API.';
COMMENT ON COLUMN raw_highscores.information             IS 'Source API metadata.';
COMMENT ON COLUMN raw_highscores.source_file_date        IS 'Business reference date of the source data snapshot.';
COMMENT ON COLUMN raw_highscores.source_file_path        IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN raw_highscores.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN raw_highscores.ingested_at             IS 'Bronze layer ingestion timestamp (UTC).';
```
