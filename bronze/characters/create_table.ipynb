```sql
USE CATALOG tibia_analytics;
USE SCHEMA bronze;
```

```sql
CREATE TABLE IF NOT EXISTS raw_characters (
  raw_characters_id STRING,
  character STRUCT<
    character: STRUCT<
      name: STRING,
      traded: BOOLEAN,
      deletion_date: STRING,
      former_names: ARRAY<STRING>,
      title: STRING,
      unlocked_titles: INT,
      sex: STRING,
      vocation: STRING,
      level: INT,
      achievement_points: INT,
      world: STRING,
      former_worlds: ARRAY<STRING>,
      residence: STRING,
      married_to: STRING,
      houses: ARRAY<STRUCT<
        houseid: INT,
        name: STRING,
        town: STRING,
        paid: STRING
      >>,
      guild: STRUCT<
        name: STRING,
        rank: STRING
      >,
      last_login: STRING,
      comment: STRING,
      position: STRING,
      account_status: STRING      
    >,
    account_badges: ARRAY<STRUCT<
      name: STRING,
      description: STRING,
      icon_url: STRING
    >>,
    achievements: ARRAY<STRUCT<
      grade: INT,
      name: STRING,
      secret: BOOLEAN
    >>,
    deaths_truncated: BOOLEAN,
    deaths: ARRAY<STRUCT<
      time: STRING,
      level: INT,
      reason: STRING,
      assists: ARRAY<STRUCT<
        name: STRING,
        player: BOOLEAN,
        traded: BOOLEAN,
        summon: STRING
      >>,
      killers: ARRAY<STRUCT<
        name: STRING,
        player: BOOLEAN,
        traded: BOOLEAN,
        summon: STRING
      >>
    >>,
    account_information: STRUCT<
      position: STRING,
      loyalty_title: STRING,
      created: STRING
    >,
    other_characters: ARRAY<STRUCT<
      name: STRING,
      position: STRING,
      world: STRING,
      status: STRING,
      main: BOOLEAN,
      traded: BOOLEAN,
      deleted: BOOLEAN
    >>
  >,
  information STRUCT<
    api: STRUCT<
      version: INT,
      release: STRING,
      commit: STRING
    >,
    timestamp: STRING,
    tibia_urls: ARRAY<STRING>,
    status: STRUCT<
      error: INT,
      http_code: INT,
      message: STRING
    >
  >,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_characters                         IS 'Raw characters data with ingestion metadata.';
COMMENT ON COLUMN raw_characters.raw_characters_id       IS 'Technical unique identifier.';
COMMENT ON COLUMN raw_characters.character               IS 'Raw characters data extracted from the source API.';
COMMENT ON COLUMN raw_characters.information             IS 'Source API metadata.';
COMMENT ON COLUMN raw_characters.source_file_date        IS 'Business reference date of the source data snapshot.';
COMMENT ON COLUMN raw_characters.source_file_path        IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN raw_characters.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN raw_characters.ingested_at             IS 'Bronze layer ingestion timestamp (UTC).';
```

```sql
CREATE TABLE IF NOT EXISTS raw_characters_attempts (
  character_id STRING,
  name STRING,
  world STRING,
  status STRING,
  source_file_date DATE,
  source_file_path STRING,
  source_file_modified_at TIMESTAMP,
  ingested_at TIMESTAMP
);
```

```sql
COMMENT ON TABLE  raw_characters_attempts                         IS 'Raw log of API fetch attempts for characters with ingestion metadata.';
COMMENT ON COLUMN raw_characters_attempts.character_id            IS 'Stable unique identifier of the character across name changes.';
COMMENT ON COLUMN raw_characters_attempts.name                    IS 'The name of the character.';
COMMENT ON COLUMN raw_characters_attempts.world                   IS 'The world of the character.';
COMMENT ON COLUMN raw_characters_attempts.status                  IS 'Status of the API attempt. Possible values: success, failed.';
COMMENT ON COLUMN raw_characters_attempts.source_file_date        IS 'Business reference date of the source data snapshot.';
COMMENT ON COLUMN raw_characters_attempts.source_file_path        IS 'Source file storage path for data lineage.';
COMMENT ON COLUMN raw_characters_attempts.source_file_modified_at IS 'Source file last modification timestamp (UTC).';
COMMENT ON COLUMN raw_characters_attempts.ingested_at             IS 'Bronze layer ingestion timestamp (UTC).';
```
